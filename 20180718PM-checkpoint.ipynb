```python
# 
import math

def fun(n):
    s = 0
    for i in range(0, n + 1):
        s += 1 / math.factorial(i)  # 1/i
    return s

print(fun(10))
fun(1000)
```

```python
# 

import math

def fun(n):
    if n == 0:
        return 1
    return 1 / math.factorial(n) + fun(n - 1)

print(fun(100))
```

```python
import time 

def run():
    while True :
        t = time.localtime()
        print("%d:%d:%d" % (t[3], t[4], t[5]), end = " \r")
        time.sleep(1)
        
run()
```

```python
def f(n):
    s = 0
    sign = 1  #  
    for i in range(1, n+1):
        s += sign * 1 / (2 * i - 1)
        sign *= -1
        
    return s

print(f(1000000))
print(f(1000000)*4)
    
```

```python
# 1

import time

def alarm(h, m):
    while True:
        cur_time = time.localtime()  # 
        print("%02d:%02d:%02d " % cur_time[3:6], end = '\r')
        time.sleep(1)
        if h == cur_time[3] and m == cur_time[4]:
            print("....")
            return

if __name__ == '__main__':
    
    hour = int(input(""))
    minute = int(input(""))
    alarm(hour, minute)
    
```

```python
# 2
import random


kinds = ['\u2660', '\u2663', '\u2665', '\u2666']

numbers = ['A'] + [str(i) for i in range(2, 11)] + list("JQK")

pokes = ['', ''] # 

for k in kinds:
    for n in numbers:
        pokes.append(k + n)

# 
random.shuffle(pokes)        
# 
input()
print("17", pokes[:17])
print("17", pokes[17:34])
print("17", pokes[34:51])
input()
print("", pokes[51:])
# print(pokes)
# print(len(pokes))
```
