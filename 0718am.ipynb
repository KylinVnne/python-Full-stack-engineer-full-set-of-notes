## 模块的文档字符串：

模块内第一个没有赋值给任何变量的字符串称为模块的文档字符串

**模块的__doc__属性：**
    
    此模块用于绑定模块的文档字符串；
    
**模块的__file__属性：**

    此属性用于记录模块对应的文件路径
    
**模块的__name__属性：**

    作用：
    
    1、记录模块名；
    
    2、用来判断是否为主模块
    
    说明：
    
    当此模块为主模块（也就是第一个运行的模块时）运行时，__name__绑定“__main__”
    
    当此模块不是模块时，而是被其他模块导入时，此时__name__绑定模块名。

```python
# 此实例为模块的文档字符串

import numpy as np

help(np)
```

```python
# 此实例是mymod3的__name__属性的用法

def test():
    pass

print("__name__属性绑定的值是：",__name__)

if __name__ == "__main__":
    print("当前mymod3.py 正在以主模块运行：")
    
else:
    print("当前mymod3.py 正在以主模块运行")
    print("我的模块名",__name__)
    
```
