```python
# 类列举
class Man:
    def __init__(self, name):
        self.name = name
        print("Initalized!")
        
    def hello(self):
        print("Hello " + self.name + "!")
        
    def goodbye(self):
        print("Good bye " + self.name + "!")
        
m = Man("Panda")
m.hello()
m.goodbye()
```

# 文件 File

**文件概念：**

文件时用于数据存储的单位，文件通常用来长期存储数据，文件中的数据是以字节为单位进行

**文件的操作流程：**

1、打开文件

2、读/写操作

3、关闭文件

注意：

任何的操作系统，一个应用程序同时打开文件的数量有最大数限制

**文件的打开函数：**

open（file， mode = ‘rt’） 用于打开一个文件，返回此文件流对象，若果打开文件失败，则会触发OSError错误。

**文件的关闭方法：**

F.close（）  # 关闭文件，释放系统资源


```python
# 文件的打开方式和关闭系统实例
try:
    f = open("./pannengxiang.txt")
    print("打开文件成功")
#     此处要进行读写操作
    f.close()
except OSError:
    print("文件打开失败")
```

```python
try:
    f = open("./myfile.txt")
    print("文件打开成功")
    
    # 此处要进行读操作
    s = f.readline() # 读取一行数据，当读取到“\n”就返回
    print("len(s) = ", len(s))
    print("s:", s)
    s = f.readline() # 读取第二行
    print("第二行：", s)
    s = f.readline()  # 读取第三行
    print("第三行：", s)
    s = f.readline()  # 只能读取空字符串（表示已经到达文件尾）
    if s == " ":
        print("——————文件尾") 
    f.close()
    
except OSError:
    print("文件打开失败")
```

```python
# 练习：写一个文件“information.txt”内部存一些文字信息
try:
    f = open("./information.txt")
    print("文件打开成功！")
    
    # 此处要进行读操作
    L = f.readlines()
    for line in L:
        print(line.strip())
    
    f.close()
except OSError:
    print("文件打开失败！")
```

```python
# 法二：
try:
    f = open("./information.txt")
    L = f.readlines()
    for line in L:
        s = line.strip()
        n, a, s = s.split() # 拆成列表
        a = int(a)
        s = int(s)
        print("name:", n, "age:",a, "score:",s)
    f.close()
except OSError:
    print("文件打开失败！")
```
