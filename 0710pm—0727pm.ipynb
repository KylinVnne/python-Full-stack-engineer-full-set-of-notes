### 集合的运算:

交集、并集、补集、子集、超集

& 交集：生成两个集合的交集

| 并集：生成两个集合的并集

- 补集：生成两个集合的补集

^ 子集：生成两个集合的对称补集，即为子集

```python

s1 = {1, 2, 3, 4}
s2 = {3, 5, 2, 7, 9}

s3 = s1&s2  # &交集：生成两个集合的交集

s4 = s1|s2  #| 并集：生成两个集合的并集

s5 = s1-s2  # - 补集：生成两个集合的补集

print(s3, s4, s5)
```

```python
s1 = {1, 2, 3}
s2 = {2, 3}
print(s1 > s2) # True s1为s2的超集
print(s1 < s2) # False s2为s1的子集
print(s1 == s2)
# “>” 判断一个集合是另一个集合的超集
# < 判断一个集合是另一个集合的子集
# == != 集合的相同/不相同
```

```python
"""
写一个函数mysum，传递一个参数x代表终止整数，打印出1 + 2 + 3 + .....+ x的和，def mysum（x）：
"""
def mysum(x):
    s = 0
    n = 1
    while n <= x:
        s = s + n
        n += 1

    return s

print(mysum(100))
print(mysum(4))

```

## 函数的参数传递：

   ### 传递方式：
   
   1-位置传参:
   
     序列传参
   
   2-关键字传参:
   
     字典关键字传参
   
   ### 1-位置传参：
    
   实际参数（实参）的对应关系与形式参数（形参）的对应关系是按位置来依此对应的
   
   **说明：**
   
   实参和形参通过位置进行传递和匹配
   
   实际参数的个数必须与形式参数的个数相同
   

```python
# 位置传参

def myfun1(a, b, c):
    print("a = ",a)
    print("b= ", b)
    print("c = ", c)
    
myfun1(1, 2, 3)
myfun1(1, 2, 3)
```

### 序列传参

序列传参是指在函数调用的过程中，用* 将序列拆解后按位置进行传递的传参方式。

说明：

序列传参时，序列拆解的位置将与形参一一对应

序列的位置信息对应相应的形参位置。

```python
# 序列传参

def myfun1(a, b, c):
    print("a = ",a)
    print("b= ", b)
    print("c = ", c)
    
s1 = [11, 22, 33]
myfun1(*s1)  # 相当于myfun1(11, 22, 33)
t1 = (44, 55, 66)
myfun1(*t1) # 星号（*）相当于拆解序列
myfun1(*"ABC")
```

```python
# 序列传参

def myfun1(a, b, c):
    print("a = ",a)
    print("b= ", b)
    print("c = ", c)
    
s1 = [11, 22, 33]
myfun1(*s1)  # 相当于myfun1(11, 22, 33)
```

### 2-关键字传参

关键字传参是指传参时，按着形参的名称给形参赋值

实参和形参按形参名进行匹配

```python
# 关键字传参

def myfun1(a, b, c):
    print("a = ",a)
    print("b= ", b)
    print("c = ", c)
    
myfun1(b=22, c=33, a=11)
myfun1(c=666, b=555, a=444)
# myfun1(c=666, b=555)
# myfun1(c=666, b=555, b=444)
```

### 字典关键字传参

是指实参为字典，将字典用"**" 拆解后再进行关键字传参的传参方式。

实例：

说明：

字典的键名和形参名必须一致

字典的键名必须为字符串

字典的键名要在形参中存在


```python
# 字典关键字传参

def myfun1(a, b, c):
    print("a = ",a)
    print("b= ", b)
    print("c = ", c)
    
d1 = {'a':100, 'b':200, 'c':300}
myfun1(**d1)  # 等同于myfun1（a = 100， b = 200， c = 300）

# d2 = {'a':100, 'b':200, 'c':300,'d':400}
# myfun1(**d2)  # 等同于myfun1（a = 100， b = 200， c = 300,）
```

## 函数的综合传参

函数的传参方式在能确定形参能唯一匹配到相应实参的情况下可以任意组合。

说明：

位置传参（序列传参）要在关键字传参（字典关键字传参）的左侧。


```python
# 函数的综合传参

def myfun1(a, b, c):
    print("a = ",a)
    print("b= ", b)
    print("c = ", c)
    
myfun1(100, *[200, 300])

myfun1(*(100,200), 300)

myfun1(*[100], 200, *(300,))

myfun1(100, c=300, b=200) # right

# myfun1(c=300, b=200, 100) # false
myfun1(100,c=300, b=200)
# myfun1(100, **{'c':300,'b':200})
```
