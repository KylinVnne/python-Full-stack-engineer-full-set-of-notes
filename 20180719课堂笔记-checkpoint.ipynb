## 异常（基础）except

**什么是错误：**

错误是指由于逻辑或语句语法等导致一个程序已无法正常执行的问题；

**什么是异常：**

异常时候程序出错时标识的一中状态；

当异常发生时，程序不会再向下执行，而转去调用此函数的地方待处理此错误并恢复为正常状态；

**异常的作用：**

用作信号，通知上层调用者有错误产生需要处理。
 
### try 语句：

**两种语法：**

    try-except语句；
    
    try-finally语句
    
**try-except语句语法：**

    try：
     
        可能触发异常的语句
        
    except  错误类型1 [as 变量1]
    
        异常处理语句1
        
    except  错误类型2 [as 变量2]
    
        异常处理语句2
            
    except （错误类型3，错误类型4，...） [as 变量3]
    
        异常处理语句3
        
    .....
    
    except:
    
        异常处理语句（其他）
        
    else：
    
        未发生异常语句
        
    finall：
    
        最终语句
        
**作用：**

尝试捕捉异常，将程序由异常状态转为正常状态并正常执行。

**示例如下：**

**try—except语句说明：**

1、as子句是用于绑定错误对象的变量。可以省略；

2、except 子句可以有一个或多个，但至少要有一个；

3、else子句最多只能有一个，也可以省略不写；

4、finally 子句最多只能有一个，也可以省略不写。

5、else子句的语句将在此try语句没有发生异常时被执行。

6、finally子句的语句在任何情况下都会被执行。

```python
# try-except语句例子：

def div_apple(n):
    print("%d个苹果您想分给几个人？"%n)
    s = input("请输入人数：")
    cnt = int(s) # 此处可能触发ValueError类型错误；
    result = n / cnt # 此处可能触发ZeroDivisionError错误；
    print("每个人分了%d个苹果"% result)
    import tensorflow
    
try:
    div_apple(10)  # 可能触发异常的调用
    print("结束分苹果！")
except ValueError as err:
    print("输入人数类型有误，已经处理并转为正常状态！")
    #     要在此处知道错误的原因。
    print("错误的原因：",err)
    
except ZeroDivisionError as err:
    print("输入人数为0，已经正常处理")
        #     要在此处知道错误的原因。
    print("错误的原因：",err)
except:  # 以上没有成功匹配的错误类型在此处都会被捕获
    print("其他的类型的错误，已经捕获并转为正常状态。")
    
print("程序正常执行并完全执行语句，程序退出！")
```

## try-finally语句：

**语法：**

try：

    可能触发异常的语句
  
finally：

    最终语句
    
**说明：**

1、finally 子句不可能省略

2、一定不存在except 子句；

**作用：**

通常用try-finally 语句来做触发异常是必须要处理的事情。无论异常是否发生，finally子句都会被执行。

**注意：**

try-finally 语句不会改变程序的（正常/异常）状态；

示列如下：

```python
# try-finally

# 以下以煎蛋为实例讲解try-finally语句的用法，必须要做的事是关闭天然气。
def fry_egg():
    print("打开天然气开关，点燃....")
    try:  
        count = int(input("请输入鸡蛋的个数："))
        print("完成煎鸡蛋，共煎了%d个鸡蛋！" % count)
    finally:
        print("关闭天然气")
    
try:
    fry_egg()
    print("煎鸡蛋成功！")
except ValueError:
    print("煎鸡蛋出现错误，已经转为正常状态！")
    
```

```python
def fry_egg():
    try:
        print("打开天然气开关，点燃....")
        try:
            count = int(input("请输入鸡蛋的个数："))
            print("完成煎鸡蛋，共煎了%d个鸡蛋！" % count)
        finally:
            print("关闭天然气")          
    except ValueError:
        print("煎蛋失败，并已经处理！")
    
fry_egg()
# print（“煎蛋成功!”）
```

## raise语句：

**作用：**

触发一个错误，让程序进入异常状态；发送错误通知给调用者。

**语法：**

raise 异常类型

或

raise 异常对象

示列见：

```python
# raise.py

def make_exception():
    print("函数开始")
#     发出一个ValueErroe错误通知给调用者
#     raise ValueError
    
    err = ValueError("这是自定义的错误数据")
    raise err  # 用err触发一个异常通知
    print("函数结束")
   
try:
    make_exception()
    print("make_exception调用结束！")
except ValueError as e:
    print("接受到ValueError类型的异常通知！")
    print("错误对象是：",e)
```

## 练习：



```python
def get_age():
    a = int(input("请输入年龄（0-140）："))
    if a > 140:
        raise ValueError("年龄超出了正常值")
    if a < 1:
        raise ValueError("年龄太小...")
    return a

try:
    
    age = get_age()
    print("用户输入的年龄是：",age)
except ValueError as err:
    print("用户输入的不是1-140的整数，获取年龄失败")
    print("原因是：",err)
    
```

## assert语句（断言语句）

**语法：**

    assert 真值表达式 ， 数据错误（通常是字符串）

**作用：**

    当真值表达式为False时，用错误数据创建一个AssertionError类型的错误，并进入异常状态

**等同于：**

    if 真值表达式 ==False

    raise AssertionError （错误数据）

    实例：
    
    assert.py

```python
# assert.py

def get_score ():
    s = int(input("请输入学生成绩："))
    assert 0 <= s <= 100, '成绩超出范围！'
    return s

try:
    score = get_score()
    print("学生的成绩为：", score)
except AssertionError as err:
    print("错误数据是：", err)
    print("获取成绩失败！")
    score = 0
    print("学生成绩为：",score)

```

## 小结：

**处理语句（四条）：**

try - except 语句：

用于接收异常通知，捕获异常；

try - finally语句：

执行必须执行的语句

raise 语句：

发送异常通知，并进入异常状态；

assert 语句：

根据条件选择性的发送AssertionError 类型的异常通知

**为什么要异常处理机制：**

在程序调用层数较深时，向主调函数传递错误信息需要层层return，返回比较麻烦，用异常处理可以较简单的传送错误的信息。

实例如下：

昨天练习讲解：



```python
# 练习1：

import time

def alarm(h, m):
    while True:
        cur_time = time.localtime()  # 得到本地时间
        print("%02d:%02d:%02d " % cur_time[3:6], end = '\r')
        time.sleep(1)
        if h == cur_time[3] and m == cur_time[4]:
            print("时间到了....")
            return

if __name__ == '__main__':
    
    hour = int(input("请输入定时的小时："))
    minute = int(input("请输入定时分钟："))
    alarm(hour, minute)
    
```
